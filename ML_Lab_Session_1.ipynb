```python
import pandas as pd
import numpy as np

def load_data():
  data = pd.read_excel("/Lab Session Data.xlsx", sheet_name= "Purchase data")

  x = data [["Candies (#)", "Mangoes (Kg)", "Milk Packets (#)"]].values
  y = data["Payment (Rs)"].values

  return x,y
def find_rank(x):
  rank = np.linalg.matrix_rank(x)
  return rank

def find_cost(x, y):
  pseudo_inverse = np.linalg.pinv(x)
  cost = np.dot(pseudo_inverse, y)
  return cost

def main():
  x,y = load_data()
  rank = find_rank(x)
  cost = find_cost(x,y)
  print("feature matrix x:")
  print(x)
  print("outpyt vector y: ")
  print(y)
  print("rank of x:", rank)
  print("cost of each product: ")
  print("cosr of one candy =", cost[0])
  print("cosr of one kg mangos= ", cost[1])
  print("cosr of one milk packet= ", cost[2])
main()
```

```python
import pandas as pd
def load_data():
  data = pd.read_excel("/Lab Session Data.xlsx", sheet_name = "Purchase data")
  return data

def classify_customers(data):
  customer_class = []

  for payment in data["Payment (Rs)"]:
    if payment > 200:
      customer_class.append("RICH")
    else:
      customer_class.append("poor")
  data["Class"] = customer_class
  return data
def main():
  data = load_data()
  classified_data = classify_customers(data)
  print("customers classfied as rich or poor: ")
  print(classified_data)
main()
```

```python
import numpy as np
import time
def calculate_mean(price):
  total = 0
  for value in price:
    total = total+value
  mean = total/len(price)
  return mean
def cal_variance(price):
  mean = calculate_mean(price)
  total = 0
  for value in price:
    total = total + (value-mean)**2
  variance = total/len(price)
  return variance
price = stock_data["Price"].values
my_mean = calculate_mean(price)
my_var = cal_variance(price)
numpy_mean = np.mean(price)
numpy_variance = np.var(price)
start = time.time()
for i in range(10):
  calculate_mean(price)
end = time.time()
my_mean_time = (end - start) / 10

start = time.time()

for i in range(10):
    np.mean(price)

end = time.time()

numpy_mean_time = (end - start) / 10

start = time.time()

for i in range(10):
    cal_variance(price)

end = time.time()

my_variance_time = (end - start) / 10

start = time.time()

for i in range(10):
    np.var(price)

end = time.time()

numpy_variance_time = (end - start) / 10

print("Mean using my function =", my_mean)
print("Mean using NumPy =", numpy_mean)

print()

print("Variance using my function =", my_var)
print("Variance using NumPy =", numpy_variance)

print()

print("Average execution time of my mean function =", my_mean_time)

print("Average execution time of NumPy mean =", numpy_mean_time)

print()

print("Average execution time of my variance function =", my_variance_time)

print("Average execution time of NumPy variance =", numpy_variance_time)
```

```python
def wednesday_mean(data):

    wednesday_data = data[data["Day"] == "Wed"]

    sample_mean = wednesday_data["Price"].mean()

    return sample_mean
population_mean = stock_data["Price"].mean()
sample_mean = wednesday_mean(stock_data)

print("popu mean =", population_mean)
print("wed sample mean =", sample_mean)

if sample_mean > population_mean:
    print("wed mean is higher than the popu mean.")
elif sample_mean < population_mean:
    print("wed mean is lower than the popu mean.")
else:
    print("both means are equal.")
```

```python
def april_mean(data):
    april_data = data[data["Month"] == "Apr"]
    average_price = april_data["Price"].mean()
    return average_price
overall_mean = stock_data["Price"].mean()
april_average = april_mean(stock_data)
print("overall avg price =", overall_mean)
print("avg price in april =", april_average)
if april_average > overall_mean:
    print("the avg stock price in April is higher than the overall avg.")
elif april_average < overall_mean:
    print("the avg stock price in April is lower than the overall avg.")
else:
    print("Both avgs are equal.")
```

```python
def loss_probability(data):
    total_days = len(data)
    loss_days = len(data[data["Chg%"] < 0])
    probability = loss_days / total_days
    return probability
loss_prob = loss_probability(stock_data)
print("prob of making a loss =", loss_prob)
```

```python
def profit_given_wednesday(data):
    wednesday_data = data[data["Day"] == "Wed"]
    total_wednesdays = len(wednesday_data)
    profit_days = len(wednesday_data[wednesday_data["Chg%"] > 0])
    probability = profit_days / total_wednesdays
    return probability
profit_probability = profit_given_wednesday(stock_data)
print("prob of making profit given wed is =", profit_probability)
```
