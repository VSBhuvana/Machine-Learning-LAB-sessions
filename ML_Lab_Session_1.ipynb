```python
import pandas as pd
import numpy as np

def load_data():
  data = pd.read_excel("/Lab Session Data.xlsx", sheet_name= "Purchase data")

  x = data [["Candies (#)", "Mangoes (Kg)", "Milk Packets (#)"]].values
  y = data["Payment (Rs)"].values

  return x,y
def find_rank(x):
  rank = np.linalg.matrix_rank(x)
  return rank

def find_cost(x, y):
  pseudo_inverse = np.linalg.pinv(x)
  cost = np.dot(pseudo_inverse, y)
  return cost

def main():
  x,y = load_data()
  rank = find_rank(x)
  cost = find_cost(x,y)
  print("feature matrix x:")
  print(x)
  print("outpyt vector y: ")
  print(y)
  print("rank of x:", rank)
  print("cost of each product: ")
  print("cosr of one candy =", cost[0])
  print("cosr of one kg mangos= ", cost[1])
  print("cosr of one milk packet= ", cost[2])
main()
```

```python
import numpy as np
import time
def calculate_mean(price):
  total = 0
  for value in price:
    total = total+value
  mean = total/len(price)
  return mean
def cal_variance(price):
  mean = calculate_mean(price)
  total = 0
  for value in price:
    total = total + (value-mean)**2
  variance = total/len(price)
  return variance
price = stock_data["Price"].values
my_mean = calculate_mean(price)
my_var = cal_variance(price)
numpy_mean = np.mean(price)
numpy_variance = np.var(price)
start = time.time()
for i in range(10):
  calculate_mean(price)
end = time.time()
my_mean_time = (end - start) / 10

start = time.time()

for i in range(10):
    np.mean(price)

end = time.time()

numpy_mean_time = (end - start) / 10

start = time.time()

for i in range(10):
    cal_variance(price)

end = time.time()

my_variance_time = (end - start) / 10

start = time.time()

for i in range(10):
    np.var(price)

end = time.time()

numpy_variance_time = (end - start) / 10

print("Mean using my function =", my_mean)
print("Mean using NumPy =", numpy_mean)

print()

print("Variance using my function =", my_var)
print("Variance using NumPy =", numpy_variance)

print()

print("Average execution time of my mean function =", my_mean_time)

print("Average execution time of NumPy mean =", numpy_mean_time)

print()

print("Average execution time of my variance function =", my_variance_time)

print("Average execution time of NumPy variance =", numpy_variance_time)
```

```python
import pandas as pd

def load_data():
    data = pd.read_excel("Lab Session Data.xlsx", sheet_name="thyroid0387_UCI")
    return data
def similarity_measure(data):
    binary_data = data.loc[:, data.isin(['t', 'f']).all()]
    vector1 = binary_data.iloc[0]
    vector2 = binary_data.iloc[1]
    f11 = 0
    f10 = 0
    f01 = 0
    f00 = 0
    for i in range(len(vector1)):

        if vector1.iloc[i] == 't' and vector2.iloc[i] == 't':
            f11 += 1

        elif vector1.iloc[i] == 't' and vector2.iloc[i] == 'f':
            f10 += 1

        elif vector1.iloc[i] == 'f' and vector2.iloc[i] == 't':
            f01 += 1

        elif vector1.iloc[i] == 'f' and vector2.iloc[i] == 'f':
            f00 += 1

    jc = f11 / (f11 + f10 + f01)
    smc = (f11 + f00) / (f11 + f10 + f01 + f00)
    return jc, smc
def main():

    data = load_data()

    jc, smc = similarity_measure(data)

    print("Jaccard Coefficient =", jc)
    print("Simple Matching Coefficient =", smc)
main()
```

```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

def load_data():
    data = pd.read_excel("Lab Session Data.xlsx", sheet_name="thyroid0387_UCI")
    return data

def cosine_similarity(data):
    encoder = LabelEncoder()

    for column in data.columns:
        if data[column].dtype == "object":
            data[column] = encoder.fit_transform(data[column].astype(str))

    vector1 = data.iloc[0].values
    vector2 = data.iloc[1].values

    dot_product = np.dot(vector1, vector2)

    magnitude1 = np.linalg.norm(vector1)
    magnitude2 = np.linalg.norm(vector2)

    cosine = dot_product / (magnitude1 * magnitude2)

    return cosine

def main():
    data = load_data()
    cosine = cosine_similarity(data)
    print("Cosine Similarity =", cosine)

main()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

def load_data():
    data = pd.read_excel("Lab Session Data.xlsx", sheet_name="thyroid0387_UCI")
    return data

def prepare_data(data):
    encoder = LabelEncoder()

    for column in data.columns:
        if data[column].dtype == "object":
            data[column] = encoder.fit_transform(data[column].astype(str))

    return data.iloc[:20]

def jaccard(v1, v2):
    f11 = f10 = f01 = 0

    for i in range(len(v1)):
        if v1[i] == 1 and v2[i] == 1:
            f11 += 1
        elif v1[i] == 1 and v2[i] != 1:
            f10 += 1
        elif v1[i] != 1 and v2[i] == 1:
            f01 += 1

    if (f11 + f10 + f01) == 0:
        return 0

    return f11 / (f11 + f10 + f01)

def smc(v1, v2):
    f11 = f10 = f01 = f00 = 0

    for i in range(len(v1)):
        if v1[i] == 1 and v2[i] == 1:
            f11 += 1
        elif v1[i] == 1 and v2[i] != 1:
            f10 += 1
        elif v1[i] != 1 and v2[i] == 1:
            f01 += 1
        else:
            f00 += 1

    return (f11 + f00) / (f11 + f10 + f01 + f00)

def cosine(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def similarity_matrix(data, method):

    n = len(data)

    matrix = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            if method == "JC":
                matrix[i][j] = jaccard(data.iloc[i].values, data.iloc[j].values)

            elif method == "SMC":
                matrix[i][j] = smc(data.iloc[i].values, data.iloc[j].values)

            else:
                matrix[i][j] = cosine(data.iloc[i].values, data.iloc[j].values)

    return matrix

def main():

    data = load_data()

    data = prepare_data(data)

    jc = similarity_matrix(data, "JC")
    smc_matrix = similarity_matrix(data, "SMC")
    cos = similarity_matrix(data, "COS")

    plt.figure(figsize=(8,6))
    sns.heatmap(jc)
    plt.title("Jaccard Coefficient")
    plt.show()

    plt.figure(figsize=(8,6))
    sns.heatmap(smc_matrix)
    plt.title("Simple Matching Coefficient")
    plt.show()

    plt.figure(figsize=(8,6))
    sns.heatmap(cos)
    plt.title("Cosine Similarity")
    plt.show()

main()
```
