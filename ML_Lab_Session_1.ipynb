```python
import pandas as pd
import numpy as np

def load_data():
  data = pd.read_excel("/Lab Session Data.xlsx", sheet_name= "Purchase data")

  x = data [["Candies (#)", "Mangoes (Kg)", "Milk Packets (#)"]].values
  y = data["Payment (Rs)"].values

  return x,y
def find_rank(x):
  rank = np.linalg.matrix_rank(x)
  return rank

def find_cost(x, y):
  pseudo_inverse = np.linalg.pinv(x)
  cost = np.dot(pseudo_inverse, y)
  return cost

def main():
  x,y = load_data()
  rank = find_rank(x)
  cost = find_cost(x,y)
  print("feature matrix x:")
  print(x)
  print("outpyt vector y: ")
  print(y)
  print("rank of x:", rank)
  print("cost of each product: ")
  print("cosr of one candy =", cost[0])
  print("cosr of one kg mangos= ", cost[1])
  print("cosr of one milk packet= ", cost[2])
main()
```

```python
import pandas as pd
def load_data():
  data = pd.read_excel("/Lab Session Data.xlsx", sheet_name = "Purchase data")
  return data

def classify_customers(data):
  customer_class = []

  for payment in data["Payment (Rs)"]:
    if payment > 200:
      customer_class.append("RICH")
    else:
      customer_class.append("poor")
  data["Class"] = customer_class
  return data
def main():
  data = load_data()
  classified_data = classify_customers(data)
  print("customers classfied as rich or poor: ")
  print(classified_data)
main()
```
